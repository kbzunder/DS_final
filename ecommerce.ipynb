```python
# group
grouped = data.groupby('customer_id', as_index=False).agg({'order_id':'nunique', 'price':'sum'})\
    .rename(columns= {'order_id':'orders'}).sort_values(by='orders', ascending=False).reset_index(drop=True)
```

```python
grouped.head()
```

```python
fig = px.scatter(grouped, x='orders', y='price')
fig.show()
```
